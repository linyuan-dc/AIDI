```python
# Question 33, Assignment 1
# firstorder partial derivatives of f(x,y)
from sympy import plot_implicit, Eq
from sympy.abc import x, y
p=plot_implicit (Eq(4*y*y*y+2*x,0) ,(x, -2, 2), (y, -2, 2),show=False, line_color='r')
p2=plot_implicit (Eq(2*y+2*x,0) ,(x, -2, 2), (y, -2, 2),show=False, line_color='b')
p.extend(p2)
p.show()
```

```python
# Question 33, Assignment 1
# 3D Graph of f(x,y)
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt

def z_function(x, y):
    return x**2+2*x*y+y**4

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

fig = plt.figure(figsize=(12,12))
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')
ax.set_title('f(x,y)=x**2+2*x*y+y**4')
ax.set_zlim3d(-0.5, 2)
ax.view_init(-0,105)
```
